### Model

```python
# Create a new Gurobi model
model = gp.Model("Beer_Distribution_Problem")

# Create variables
inputs_1 = model.addVars(brewing, packaging, lb=0, vtype=GRB.CONTINUOUS, name="inputs_1")
inputs_2 = model.addVars(packaging, demand_points, lb=0, vtype=GRB.CONTINUOUS, name="inputs_2")

# Objective function
objective_expr = gp.quicksum(inputs_1[i, j] * brew_to_pack_shipping_costs[brewing.index(i)][packaging.index(j)]
                             for i in brewing for j in packaging) + \
                  gp.quicksum(inputs_2[j, k] * pack_to_demand_shipping_costs[container][demand_points.index(k) + 1][packaging.index(j) + 1]
                             for j in packaging for k in demand_points for container in pack_to_demand_shipping_costs)
model.setObjective(objective_expr, GRB.MINIMIZE)

# Demand constraints
for c in commodities:
    for k in total_dp_demand.keys():
        required_qty = demand[c][k] if k in demand[c] else 0
        model.addConstr(gp.quicksum(inputs_2[j, k] for j in packaging) >= required_qty,
                        f"Demand_Constraint_Commodity{c}_DemandPoint{k}")

# Brewery constraints
for i in brewing:
    model.addConstr(gp.quicksum(inputs_1[i, j] for j in packaging) <= brewing_maximum[i],
                    f"Maximum_Brewing{i}")
    model.addConstr(gp.quicksum(inputs_1[i, j] for j in packaging) >= brewing_minimum[i],
                    f"Minimum_Brewing{i}")

# Packaging constraints
for j in packaging:
    model.addConstr(gp.quicksum(inputs_2[j, k] for k in demand_points) <= packaging_maximum[j],
                    f"Maximum_Packaging{j}")
    model.addConstr(gp.quicksum(inputs_2[j, k] for k in demand_points) >= packaging_minimum[j],
                    f"Minimum_Packaging{j}")

# Packaging output constraint
for j in packaging:
    model.addConstr(gp.quicksum(inputs_2[j, k] for k in demand_points) == gp.quicksum(inputs_1[i, j] for i in brewing),
                    f"Packaging_Output{j}")

# Brewery liquid commodity and container commodity constraints
for i in brewing:
    for c in commodities:
        required_liquid = commodities[c]["Liquid"]
        max_qty = liquid_constraints[required_liquid]["Max Qty"]
        if breweries_liquids_config[i][required_liquid] == 1:
            model.addConstr(gp.quicksum(inputs_1[i, j] * commodities[c]["Container"] for j in packaging) <= max_qty,
                            f"Liquid_Constraint_Brewery{required_liquid}_{i}_Commodity{c}")
        else:
            model.addConstr(0 <= 0, f"Liquid_Constraint_Brewery{required_liquid}_{i}_Commodity{c}")

# Solve the model
model.optimize()

```

### Optimal Solution & Sensitivity Report

```python
# Check if the optimization was successful
if model.Status == GRB.OPTIMAL:
    print("Optimal solution found.")
    # Print objective value
    print("Objective value:", round(model.ObjVal))

    # Print variable values
    print("\nVariable values:")
    for v in model.getVars():
        print(v.varName, "=", round(v.x))

    # Sensitivity report
    print("\nSensitivity report:")

    # Shadow prices (dual values) for demand constraints
    print("\nShadow prices (dual values) for demand constraints:")
    for c in commodities:
        for k in total_dp_demand.keys():
            print(f"Commodity{c}_DemandPoint{k}:", model.getConstrByName(f"Demand_Constraint_Commodity{c}_DemandPoint{k}").Pi)

    # Shadow prices for brewery constraints
    print("\nShadow prices for brewery constraints:")
    for i in brewing:
        print(f"Maximum_Brewing{i}:", model.getConstrByName(f"Maximum_Brewing{i}").Pi)
        print(f"Minimum_Brewing{i}:", model.getConstrByName(f"Minimum_Brewing{i}").Pi)

    # Shadow prices for packaging constraints
    print("\nShadow prices for packaging constraints:")
    for j in packaging:
        print(f"Maximum_Packaging{j}:", model.getConstrByName(f"Maximum_Packaging{j}").Pi)
        print(f"Minimum_Packaging{j}:", model.getConstrByName(f"Minimum_Packaging{j}").Pi)

else:
    print("Optimization was not successful.")

# Aggregating brewing and packaging totals
brewing_outputs = {i: sum(inputs_1[i, j].x for j in packaging) for i in brewing}
packaging_inputs = {j: sum(inputs_2[j, k].x for k in demand_points) for j in packaging}
packaging_outputs = {j: sum(inputs_2[j, k].x for k in demand_points) for j in packaging}

print("\n")
print("Brewing Outputs:")
for i, output in brewing_outputs.items():
    print(f"{i}: {output}")

print("Packaging Inputs:")
for j, input_ in packaging_inputs.items():
    print(f"{j}: {input_}")

print("Packaging Outputs:")
for j, output in packaging_outputs.items():
    print(f"{j}: {output}")

# Total demand and shipping costs
total_demand = sum(total_dp_demand.values())
total_shipping_costs = round(model.ObjVal)

print("\n")
print("Total Demand:", total_demand)
print("Total Shipping Costs:", total_shipping_costs)
```
